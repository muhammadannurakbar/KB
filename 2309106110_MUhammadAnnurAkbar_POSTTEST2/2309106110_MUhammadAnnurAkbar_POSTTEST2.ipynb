## Jumlah label atribut objek

## Visualisasi data

### Subtask:
Pilih dan visualisasikan dataset menggunakan grafik yang paling sesuai (misalnya, distribusi target variabel).


**Reasoning**:
Impor pustaka visualisasi yang diperlukan, buat grafik batang untuk variabel target 'Diabetes_012', atur label dan judul, lalu tampilkan grafik.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(x='Diabetes_012', data=df)
plt.title('Distribusi Variabel Target: Diabetes_012')
plt.xlabel('Kategori Diabetes')
plt.ylabel('Jumlah Data')
plt.xticks(ticks=[0, 1, 2], labels=['Tidak Diabetes', 'Pra-Diabetes', 'Diabetes'])
plt.show()
```

## Heatmap korelasi

### Subtask:
Hitung dan visualisasikan matriks korelasi menggunakan heatmap.


**Reasoning**:
Hitung matriks korelasi dan visualisasikan menggunakan heatmap.



```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Heatmap Matriks Korelasi')
plt.show()
```
